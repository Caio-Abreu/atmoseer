```python
import os
import pandas as pd

def read_parquet_files(dir_path):
    # Initialize an empty list to store DataFrames
    dfs = []

    # Iterate over all files and directories in dir_path
    for entry in os.scandir(dir_path):
        print(f"Current entry: {entry}")
        if entry.is_file() and entry.name.endswith('.parquet'):
            dfs.append(entry.path)
        elif entry.is_dir():
            # If the entry is a directory, recursively call the function on the subdirectory
            sub_dir_path = os.path.join(dir_path, entry.name)
            print(f"Entering directory {sub_dir_path}")
            dfs.extend(read_parquet_files(sub_dir_path))

    return dfs

# Example usage: recursively read all parquet files in the directory "data"
data_dir = "../data/ws/alertario"
filenames = read_parquet_files(data_dir)
print(len(filenames))
```

```python
df_all = None
print(f"Going to concat {len(filenames)} files...")
for filename in filenames:
    print(f"Filename: {filename}")
    df = pd.read_parquet(filename)
    if df_all is None:
        df_all = df
    else:
        df_all = pd.concat([df_all, df], ignore_index=True)
df_all.shape
```

```python
df_all.to_parquet("../data/ws/alertario/ws_alertario.parquet.gzip", compression='gzip', index = False)
```
