```python
import pandas as pd
import missingno as msno

# Create a missingno object
msno.matrix(df_alto_da_boa_vista)

# Plot the missing values distribution
msno.bar(df_alto_da_boa_vista)

# Plot the missing values by column
msno.heatmap(df_alto_da_boa_vista)
```
